```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv("Absenteeism-data.csv") #importing the data into a pandas dataframe
```

```python
raw_csv_data
```

```python
df = raw_csv_data.copy() # creating a copy of the original dataframe
```

```python
df.info()
```

## Drop ID

```python
# Drop the ID column because we are not concerned with the indivduals.
# This project focuses more on the reasons for people being absent and not the people who are actually absent. 


df = df.drop(['ID'],axis=1)
```

```python
df
```

##  Create dummies for reasons for absence column

```python
print(sorted(df['Reason for Absence'].unique())) #identifying the unique reasons for absences.
```

```python
# creating the dummy variables
reason_columns = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_columns
```

```python
#checking for missing values

reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns['check'].sum(axis=0) #sum of values is 700 which is the number of observations we have
```

```python
reason_columns = reason_columns.drop(['check'],axis=1) #drop check column
reason_columns
```

```python
# recreating the dummy column this time dropping the first variable to avoid multicollinearity. 

reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

## Group the Reasons for Absence:

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
df = df.drop(['Reason for Absence'],axis=1) #dropping the reason of absence column from the original dataset
df
```

```python
reason_columns.loc[:,'1':'14']
```

```python
# grouping the reasons into 4 categories.
# 1-14 -> Sickness and/or diseases
# 15-17 -> Pregnancy
# 18-21 -> Poisoning 
# 22-28 -> Light reasons for absence e.g Dental consultation, Patient follow-up

reason_type_1 = reason_columns.loc[:,'1':'14']
reason_type_2 = reason_columns.loc[:,'15':'17']
reason_type_3 = reason_columns.loc[:,'18':'21']
reason_type_4 = reason_columns.loc[:,'22':]
reason_type_4.head()
```

```python
# use the .max() method to retreive the highest value in each row 

reason_type_1 = reason_columns.loc[:,'1':'14'].max(axis=1)
reason_type_2 = reason_columns.loc[:,'15':'17'].max(axis=1)
reason_type_3 = reason_columns.loc[:,'18':'21'].max(axis=1)
reason_type_4 = reason_columns.loc[:,'22':].max(axis=1)
reason_type_4.head()
```

## Concatenate Column Values

```python
df.head()
```

```python
# adding the reason dummies to the dataframe

df = pd.concat([df, reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df.head()
```

```python
df.columns.values
```

```python
#change the column names

column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2','Reason_3','Reason_4']
```

```python
df.columns = column_names
df.head()
```

## Reorder Columns

```python
#reordering the columns

column_names_reordered = [ 'Reason_1', 'Reason_2','Reason_3','Reason_4',
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

## Create a Checkpoint

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod.head()
```

## 'Date':

```python
# date column is being read as a string by python

type(df_reason_mod['Date'][0])
```

```python
# converting date column from string to datetime

df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

```python
type(df_reason_mod['Date'].head()[0])
```

```python
# date column has now been converted to datetime

df_reason_mod.info()
```

## Extract the Month Value:

```python
df_reason_mod['Date'][0]
```

```python
# extracting month from the first date

df_reason_mod['Date'][0].month
```

```python
# extracting all months from the dataset

list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
list_months
```

```python
len(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

## Extracting the Day of the Week:

```python
# create a function to extract day of the week

def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
# add the day of the week into the dataset

df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
# dropping the date column

df_reason_mod=df_reason_mod.drop(columns='Date')
```

```python
df_reason_mod.columns
```

```python
# rearranging the columns

df_reason_mod = df_reason_mod[['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
        'Month Value', 'Day of the Week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']]
```

```python
df_reason_date_mod = df_reason_mod.copy()
```

```python
df_reason_date_mod.head()
```

## Transforming the education column

```python
# there are 4 unique values for the Education column.They represent the highest level of education for each individual
# 1 -> highschool
# 2 -> graduate
# 3 -> post graduate
# 4 -> masters or doctorate

df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts() #distribution of the education column in the data
```

```python
# reducing education to 2 groups
# 0 -> high school
# 1 -> anything beyond high school

df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts()
```

```python
# create a final checkpoint for the preprocessed data

df_preprocessed = df_reason_date_mod.copy()
```

```python
#save preprocessed data in csv

df_preprocessed.to_csv('Absenteeism_preprocessed.csv', index=False)
```
